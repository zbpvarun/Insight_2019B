```python
#This table has 4 rows per hospital to give measures for 4 different kinds of procedures. Let us focus
#on just one kind of measure so we get one row per hospital.
location_df = df2[df2['Payment measure ID']=='PAYM_30_HF']
#Verify that the number of hospitals in both tables is the same:
np.shape(df1)[0]==np.shape(location_df)[0]
```

```python
location_df['Location']
```

```python
#For whatever reason, the location data in the downloaded table has the entire address instead of just
#the latitude and longitude. We need to extract that data from the address.
def get_location(row):
    temp = row['Location'].split('(')[-1]
    if temp == row['Location']:
        row['Latitude'] = None
        row['Longitude'] = None
        return row
    else:
        temp2 = temp.split(',')[0]
        row['Latitude'] = temp2
        #Split at the space:
        temp2 = temp.split(' ')[1]
        #Remove the last bracket:
        temp2 = temp2[:-1]
        row['Longitude'] = temp2
        return row

location_df = location_df.apply(get_location,1)
select_cols = ['Provider ID','Hospital name','State','ZIP Code','Latitude','Longitude']
location_df = location_df[select_cols]
location_df.head()
```

```python
#I need a function to compute the distance in miles between two points given their latitude and longitude data:
#To do this, I will use the Haversine function calculation:

def get_distance_haversine(lat1,lat2,long1,long2):
    import math
    del_lat = math.radians(lat1 - lat2)
    del_long = math.radians(long1 - long2)
    lat1 = math.radians(lat1)
    lat2 = math.radians(lat2)
    
    R = 6371 #Average radius of the earth in km
    a = (math.sin(del_lat/2))**2 + math.cos(lat1)*math.cos(lat2)*((math.sin(del_long/2))**2)
    c = 2*math.atan2(np.sqrt(a),np.sqrt(1 - a))
    dist = R*c #This is the distance in km
    dist_mi = dist/1.609
    return dist_mi

get_distance_haversine(38.898556,38.897147,-77.037852,-77.043934)
```

```python
#So other than the Hip and Knee Replacement score, all other features have only 2 missing values.
#Let us build 2 models, one with the hip and knee replacement and one without.

features1 = ['Heart Failure Death Score',
             'Heart Attack Death Score',
             'Pneumonia death Score',
             'Stroke death Score',
             'Blood clots post surgery Score',
             'Post surgery Sepsis Score',
             'Postop Respiratory Failure Score',
             'Postop Dialysis need Score',
             'Surgical Complication Deaths Score',
             'Hip and Knee Replacement Score']

features2 = ['Heart Failure Death Score',
             'Heart Attack Death Score',
             'Pneumonia death Score',
             'Stroke death Score',
             'Blood clots post surgery Score',
             'Post surgery Sepsis Score',
             'Postop Respiratory Failure Score',
             'Postop Dialysis need Score',
             'Surgical Complication Deaths Score',
             'Heart Failure Death Upper',
             'Heart Attack Death Upper',
             'Pneumonia death Upper',
             'Stroke death Upper',
             'Blood clots post surgery Upper',
             'Post surgery Sepsis Upper',
             'Postop Respiratory Failure Upper',
             'Postop Dialysis need Upper',
             'Surgical Complication Deaths Upper',
             'Heart Failure Death Lower',
             'Heart Attack Death Lower',
             'Pneumonia death Lower',
             'Stroke death Lower',
             'Blood clots post surgery Lower',
             'Post surgery Sepsis Lower',
             'Postop Respiratory Failure Lower',
             'Postop Dialysis need Lower',
             'Surgical Complication Deaths Lower']

X_data1 = model_test_df[features1].copy()
X_data2 = model_test_df[features2].copy()

X_data1 = X_data1.astype(float)
X_data2 = X_data2.astype(float)

for column in X_data1.columns:
    X_data1[column].fillna(value=np.nanmean(X_data1[column]),inplace=True)

for column in X_data2.columns:
    X_data2[column].fillna(value=np.nanmean(X_data2[column]),inplace=True)
    
y_data = model_test_df['Review Score'].copy()

from sklearn.model_selection import train_test_split

X_train1, X_test1, y_train1, y_test1 = train_test_split(X_data1,y_data,random_state=0)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_data2,y_data,random_state=0)
```
